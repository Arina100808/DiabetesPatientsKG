```python
patient_list = []
individual_names = []
original_patient = []

for Patient in Patient.instances():
    original_patient.append(Patient)

for i in range(100):
    individual_names.append(f"Patient{i+1}")

def create_random_patient():
    for name in individual_names:
        Patient = onto.Patient
        name = Patient(f"{name}")

    for Patient in Patient.instances():
        patient_list.append(Patient)

create_random_patient()
patient_list = [elem for elem in patient_list if elem not in original_patient]
print(patient_list)
```

```python
from datetime import datetime, timedelta

def generate_random_datetime():
    
    year = random.randint(2005, 2025)
    month = random.randint(1, 12)
    if month == 2:
        day = random.randint(1, 28)
    elif month in [1, 3, 5, 7, 8, 10, 12]:
        day = random.randint(1, 31)
    else:
        day = random.randint(1, 30)

    random_datetime = datetime(year, month, day)
    trtime = random_datetime.strftime("%Y-%m-%d")
    datetype = datetime.strptime(trtime, "%Y-%m-%d")

    return datetype
```

```python
# BMI
def randomize_BMI():
    for name in patient_list:
        name_string = str(name).split('.')[-1]
        measurement_date = generate_random_datetime()
        person_BMI_measurement = f"{name_string}_measure_BMI_{measurement_date.year}{measurement_date.month}{measurement_date.day}"
        person_BMI_measurement = BMI(f'{person_BMI_measurement}')
        person_BMI_measurement.hasBMIValue = random.randint(10, 40)
        #person_BMI_measurement.hasMeasurementDate = measurement_date
        name.hasMedicalMeasurement.append(person_BMI_measurement)
```

```python
# Debt

def randomize_Debt():
    random_patient_list = []
    patient_list_copy =  patient_list.copy()
    # define 1/3 patients in the dataset has debt
    for i in range(round(len(patient_list_copy)/3)):
        random_patient = random.choice(patient_list_copy)
        random_patient_list.append(random_patient)
        patient_list_copy.remove(random_patient)
    for name in random_patient_list:
            name.hasDebt.append(debt)
```

```python
# Gender

def randomize_Gender():
    random_patient_list = []
    patient_list_copy =  patient_list.copy()
    # Generate 50-50 Gender in the dataset
    # assign gender_male:
    for i in range(round(len(patient_list_copy)/2)):
        random_patient = random.choice(patient_list_copy)
        random_patient_list.append(random_patient)
        patient_list_copy.remove(random_patient)
    for name in random_patient_list:
        name.hasGender = gender_male
    female_list = [elem for elem in patient_list if elem not in random_patient_list]
    # assign gender_female:
    for name in female_list:
        name.hasGender = gender_female
```

```python
# SmokingHabit

def randomize_SmokingHabit():
    random_patient_list = []
    patient_list_copy =  patient_list.copy()
    # Generate smoking-patients based on 1/3 of data
    for i in range(round(len(patient_list_copy)/3)):
        random_patient = random.choice(patient_list_copy)
        random_patient_list.append(random_patient)
        patient_list_copy.remove(random_patient)
    for name in random_patient_list:
        name_string = str(name).split('.')[-1]
        person_SmokingHabit = f"{name_string}_smoking_habit"
        person_SmokingHabit = SmokingHabit(f'{person_SmokingHabit}')
        person_SmokingHabit.cigarettesPerDay = random.randint(0, 30)
        name.hasSmokingHabit = person_SmokingHabit
    non_smoking_list = [elem for elem in patient_list if elem not in random_patient_list]
    # Generate non-smoking patients:
    for name in non_smoking_list:
        name_string = str(name).split('.')[-1]
        person_SmokingHabit = f"{name_string}_smoking_habit"
        person_SmokingHabit = SmokingHabit(f'{person_SmokingHabit}')
        person_SmokingHabit.cigarettesPerDay = 0
        name.hasSmokingHabit = person_SmokingHabit
```

```python
# Value
Value_Classes = [
    Excitement,
    BeingWellRespected,
    Family,
    Friends,
    FunAndEnjoyment,
    Security,
    SelfFulfillment,
    SelfRespect,
    SenseOfAccomplishment,
    SenseOfBelonging,
    WarmRelationships,
    Work]

def randomize_Value():
    
    number_values = len(Value_Classes)

    for name in patient_list:
        name_string = str(name).split('.')[-1]
        Value_Classes_copy = Value_Classes.copy()
        randomization_time = random.randint(3, number_values)
        person_Values = []
        for i in range(randomization_time):
            Value = random.choice(Value_Classes_copy)
            Value_Classes_copy.remove(Value)
            class_name = f'{Value}'
            class_name = class_name.split(".")[-1]
            person_Value = f"{name_string}_{class_name}"
            person_Value = Value(person_Value)
            name.hasValue.append(person_Value)
            person_Values.append(person_Value)
        for i in range(len(person_Values)):
            if i < (len(person_Values)-1):
                person_Values[i].prioritizedOver.append(person_Values[i+1])
                for x in person_Values[i].prioritizedOver:
                    print(f"{person_Values[i]} prioritizedOver {x}")
```
